```python
import csv
import numpy as np 
import pandas as pd
```

```python
file_name = 'UCI_wine.csv'
```

```python
DataSet = pd.read_csv(file_name)
```

```python
DataSet.dtypes
```

```python
if 'id' in DataSet:
    print('drop id')
    DataSet = DataSet.drop(['id'],axis = 1)
```

```python
DataSet['class'] = DataSet['class'].astype('category')
```

```python
DataSet.dtypes
```

```python
Dict_Categroy_code = dict( enumerate(DataSet['class'].cat.categories))
```

```python
Dict_Categroy_code
```

```python
DataSet['class'] = DataSet['class'].cat.codes
```

```python
Result_Data = pd.DataFrame()
for feature in DataSet:
    if feature == 'class': continue
    #print(feature)
    #print(DataSet[feature].max())
    #print(DataSet[feature].apply(lambda x : x/DataSet[feature].max()))
    Result_Data[feature] = DataSet[feature].apply(lambda x : x/DataSet[feature].max())
#print(Result_Data)
```

```python
Result_Data = pd.concat([Result_Data,DataSet['class']],axis = 1)
```

```python
Result_Data
```

```python
outfile_name = file_name.replace('.csv','')+'_Normalized.csv'
```

```python
Result_Data.to_csv(outfile_name,index = False)
```
